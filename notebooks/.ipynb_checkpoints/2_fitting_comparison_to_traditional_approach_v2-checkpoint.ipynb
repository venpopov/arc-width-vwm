```python
def extract_posterior_means(df_log_posteriors):
    """
    Extracts posterior means for 'c' and 'kappa' by combining fixed effects 
    (population means) and random effects (subject deviations), then 
    transforming from log scale to native scale.

    Logic based on brms naming convention:
    Subject_Log_Param = b_Param_Intercept (Population) + r_id__Param[i,Intercept] (Deviation)
    Subject_Native_Param = exp(Subject_Log_Param)
    """
    import re
    
    # Helper function for natural sorting (ensures subject 10 comes after subject 2)
    def natural_sort_key(s):
        return [int(text) if text.isdigit() else text.lower() for text in re.split('([0-9]+)', s)]

    # Dictionary to map the target parameter to the specific column names in df
    # Key: output name, Value: (Population_Col, Deviation_Col_Pattern)
    param_map = {
        'c': ('b_c_Intercept', 'r_id__c['),
        'kappa': ('b_kappa_Intercept', 'r_id__kappa[')
    }

    results = {}

    for param, (pop_col, dev_pattern) in param_map.items():
        # 1. Identify the Population Mean Column (Fixed Effect)
        if pop_col not in df_log_posteriors.columns:
            print(f"Error: Population column {pop_col} not found.")
            continue
            
        # 2. Identify and Sort Subject Deviation Columns (Random Effects)
        # We look for columns containing the pattern (e.g., 'r_id__c[')
        # We assume the format is like 'r_id__c[1,Intercept]'
        dev_cols = [col for col in df_log_posteriors.columns if dev_pattern in col]
        dev_cols = sorted(dev_cols, key=natural_sort_key)
        
        if not dev_cols:
            print(f"Error: No deviation columns found for {param}.")
            continue

        print(f"Processing {param}: Found {len(dev_cols)} subjects.")

        # 3. Calculation: (Population + Deviation) -> Exp -> Mean
        # We do this vectorized over the dataframe for efficiency
        
        # Get Population samples (Vector: n_samples)
        pop_samples = df_log_posteriors[pop_col].values
        
        # Get Deviation samples (Matrix: n_samples x n_subjects)
        dev_samples = df_log_posteriors[dev_cols].values
        
        # Broadcast summation: Add population vector to every subject column
        # reshape(-1, 1) ensures the population vector aligns with the matrix columns
        total_log_samples = pop_samples.reshape(-1, 1) + dev_samples
        
        # 4. Transform to Native Scale
        # Apply exp() to get values back to the original physical scale (e.g., c ~ 2.0, kappa ~ 4.0)
        native_samples = np.exp(total_log_samples)
        
        # 5. Calculate Posterior Mean for each subject
        # Average across the MCMC chain (axis 0)
        subject_means = native_samples.mean(axis=0)
        
        results[f'{param}_mean'] = subject_means

    # Create the final DataFrame
    df_result = pd.DataFrame(results)
    return df_result

df_posterior_bmm = extract_posterior_means(df_log_posteriors)
df_prior_bmm = np.exp(df_true_log_parms.T.set_axis(['c_mean', 'kappa_mean'], axis=1))
```

### plot

```python
def plot_parameter_correlation(df1, df2, xlabel="prior", ylabel="posterior", figsize_per_subplot=(6, 5)):
    """
    Create scatter plots with linear regression lines for common parameters between two DataFrames.

    Parameters:
    -----------
    df1 : pandas.DataFrame
        First DataFrame containing parameter estimates with columns like 'c_mean', 'kappa_mean', etc.
    df2 : pandas.DataFrame
        Second DataFrame containing parameter estimates with columns like 'c_mean', 'kappa_mean', etc.
    figsize_per_subplot : tuple, optional
        Figure size for each subplot, default is (6, 5)

    Returns:
    --------
    matplotlib.figure.Figure
        The figure containing all correlation plots
    """
    import pandas as pd
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    
    # Find common columns between the two DataFrames
    common_columns = list(set(df1.columns) & set(df2.columns))
    
    if not common_columns:
        print("No common columns found between the two DataFrames.")
        return None
    
    # Calculate number of subplots needed
    n_params = len(common_columns)
    n_cols = 2  # Number of columns in the subplot grid
    n_rows = int(np.ceil(n_params / n_cols))
    
    # Create figure and subplots
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(figsize_per_subplot[0] * n_cols, 
                                                      figsize_per_subplot[1] * n_rows))
    
    # Handle case where there's only one subplot
    if n_params == 1:
        axes = [axes]
    elif n_params == 2:
        axes = axes.flatten()
    else:
        axes = axes.flatten() if n_rows * n_cols > 1 else [axes]
    
    # Create correlation plots for each common parameter
    for i, param in enumerate(common_columns):
        ax = axes[i]
        
        # Prepare data for plotting
        data_to_plot = pd.DataFrame({
            f'{param}_x': df1[param].values,
            f'{param}_y': df2[param].values
        })
        
        # Create scatter plot with regression line using seaborn
        sns.regplot(data=data_to_plot, 
                   x=f'{param}_x', 
                   y=f'{param}_y', 
                   ax=ax, 
                   scatter_kws={'alpha': 0.6})
        
        # Add title and labels
        ax.set_title(f'Correlation: {param}')
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        
        # Add correlation coefficient to the plot
        corr_coef = data_to_plot[f'{param}_x'].corr(data_to_plot[f'{param}_y'])
        ax.text(0.05, 0.95, f'r = {corr_coef:.3f}', 
                transform=ax.transAxes, 
                verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Hide empty subplots if any
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    return fig
```

```python
fig = plot_parameter_correlation(df_posterior_bmm, df_posterior_bayesflow, "bmm", "bayesflow")
```
